### Plots the network topology.
This example contains:
* Load a network.
* Plot network topology.
* Plot node IDs.
* Plot links IDs.
* Plot node indices.
* Plot link indices.
* Plot specific nodes ID
* Plot specific nodes index
* Plot specific links ID
* Plot specific links index
* Highlight specific links.
* Highlight specific nodes.
* Plot only links.
* Plot only nodes.
* Hide legend.
* Unload library.
      

```python
from epyt import epanet
# Load a network
d = epanet('Net1.inp')
```

```python
# Plot network topology
d.plot()
```

```python
# Plot nodes IDs
d.plot(nodesID=True)
```

```python
# Plot links IDs
d.plot(linksID=True)
```

```python
# Plot node indices
d.plot(nodesindex=True)
```

```python
# Plot link indices
d.plot(linksindex=True)
```

```python
# Plot specific nodes ID
d.plot(nodesID=['10', '11', '9'])
```

```python
# Plot specific nodes index
d.plot(nodesindex=[1, 2, 10])
```

```python
# Plot specific links ID
d.plot(linksID=['111', '122', '9'])
```

```python
# Plot specific links index
d.plot(linksindex=[1, 2, 10])
```

```python
# Highlight specific links
d.plot(highlightlink='10')
```

```python
# Highlight specific nodes 
d.plot(highlightnode=['10', '32'])
```

```python
# Plot only nodes
d.plot(point=True)
```

```python
# Plot only links
d.plot(line=True)
```

```python
# Hide legend
d.plot(legend=False)
```

```python
# Unload library
d.unload()
```
